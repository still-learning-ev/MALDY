```python
import nltk
import json
from nltk.tokenize import punkt
import string
from nltk.corpus import stopwords
import pandas as pd
import csv
```

```python
#this will convert the json data in single line txt data also reducing the size of file

with open('C:\\Users\\zeeshan lone\\Desktop\\malware report\\report2.json') as f:
    data = json.load(f)

my_string = json.dumps(data)
string = my_string.replace('\r', '').replace('\n', '')

with open('C:\\Users\\zeeshan lone\\Desktop\\malware report\\report_text.txt', 'w') as g:
    g.write(string)
```

```python
#now read the txt file and remove all of the stray punctions from the txt file

with open('C:\\Users\\zeeshan lone\\Desktop\\malware report\\report_text.txt') as re:
    raw_data = re.read()
import string
S_P=string.punctuation
def remove_punctuation(data_ne):
    text_non_punctuation="".join([char for char in data_ne if char not in S_P])
    return text_non_punctuation

token_list = nltk.word_tokenize(raw_data)
#token_list_1 = list(filter(lambda token : punkt.PunktToken(token).is_non_punct,token_list))
token_list_1 = list(filter(lambda token : remove_punctuation(token),token_list))
token_list_2 = [word.lower() for word in token_list_1]


```

```python
#remove stopwords
token_list_3 = list(filter(lambda token: token not in stopwords.words("english"),token_list_2))
```

```python
new_string = ' '.join(token_list_3)
print(len(new_string))
new_string = new_string.replace('\r', '').replace('\n', ' ')

print(new_string)
```
